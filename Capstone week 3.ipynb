Scraping data

```python
# Wiki toronto info 
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

read into pandas data frame

```python
#define dataframe columns

dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
df.rename(columns={'Postal Code':'Postcode'},inplace=True)
```

```python
column_names = ['Postcode','Borough', 'Neighborhood',] 
neighborhoods = pd.DataFrame(columns=column_names)
neighborhoods
```

cleaning data

```python
# finding missing data 
missing_data = df.isnull()
missing_data.head(5)
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
# Dropping the rows where Borough is 'Not assigned'
NotAssigned = df[df['Borough']=='Not assigned'].index
df.drop(NotAssigned, inplace=True)
df.head()
```

```python
# Combining the neighbourhoods with same Postalcode
df1 = df.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df1.reset_index(inplace=True)

df1.head()
```

```python
# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df1['Neighbourhood'] = np.where(df1['Neighbourhood'] == 'Not assigned',df1['Borough'], df1['Neighbourhood'])

df1
```

```python
#confirm shape
df1.shape
```

Latitudes and longitudes of neighbourhoods in Canada

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
# merging 
lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df1,lat_lon,on='Postcode')
df3.head()
```

```python
# getting boroughs that have toronto 
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4.head()
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

## Using KMeans to cluster the neighbourhoods

```python
k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4. head(10)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
